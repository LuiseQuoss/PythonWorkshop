### Struktur eines Jupyter Notebooks
Ein Jupyter Notebook hat Zellen mit verschiedenen Inhalten, die du mit dem Plus-Symbol einfügen kannst:
* __strukturierter Text__ (auch "Markdown" genannt)      - grüne Markierung
* und __Code__ (erkennbar an dem In [1]:) vor der Zelle  - blaue Markierung

Um den Typ des Inhalts einer Zelle zu wechseln:
* Code: drücke y oder wähle im Dropdown-Menü Code
* Text: drücke m oder wähle im Dropdown-Menü Markdown

In den Code-Zellen kannst du Python-Code schreiben und direkt ausführen, indem du __Strg + Enter__ drückst. Dein Text in den Markdown-Zellen wird über die gleiche Tastenkombination formatiert.


Falls du ein paar Anregungen für die Formatierung deiner Markdown-Zellen möchtest, kannst du bspw. gerne hier schauen: https://www.tutorialspoint.com/jupyter/jupyter_notebook_markdown_cells.htm
